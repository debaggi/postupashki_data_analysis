структура на выходе: список словарей



````python
import requests

URL = 'https://api.hh.ru/vacancies/'

SPECIALTIES = ['Аналитик данных', 'Reels maker', 'Психолог', 'Экономист', 'Системный архитектор']
AREA_ID = 1
PER_PAGE = 100


def fetch_vacancy_by_specialty(specialty):
  params = {'text':specialty,
            'area':AREA_ID,
            'page':0,
            'per_page':PER_PAGE
            }
  response = requests.get(URL, params)

  response.raise_for_status()

  data = response.json()
  result = [item['id'] for item in data.get('items', [])]
  return result


def main():
  vacancies = {}

  for specialty in SPECIALTIES:
    try:
      print(f'\nначалась обработка специальности {specialty}')
      ids = fetch_vacancy_by_specialty(specialty)
      vacancies[specialty] = ids
      print(f'✅ найдено вакансий: {len(ids)}')

    except requests.RequestException as e:
      print(f'❌ Ошибка при запросе вакансий для "{specialty}" : {e}')
    except Exception as e:
      print(f'⚠️ Неизвестная ошибка для "{specialty}": {e}')

  print(f'\nсписки ID вакансий по специальностям \n{vacancies}')


if __name__== "__main__":
  main()
````

````python
# поиск вакансий (со всеми характристиками вакансий) по списку специальностей
# код разделен на 2 части:
  # 1. обработка первой страницы с вакансиями
  # 2. обработка последующих страниц

# на входе:  URL - конечная точка,
#            params - параметры запроса

# на выходе: dataframe - атрибуты вакансий

SPECIALTIES = ['Аналитик данных', 'Reels maker', 'Психолог','Экономист','Системный архитектор']
URL = 'https://api.hh.ru/vacancies/'

list_df = []

# обработка первой страницы
# -------------------------
for specialty in SPECIALTIES:
  query = requests.get(url=URL,
                       params={
                           'text':specialty,
                           'area':1,
                           'page':0,
                           'per_page':100},
                       )

  if query.status_code != 200:
    continue


  page_vacancies_df = pd.json_normalize(query.json()['items'])
  page_vacancies_df["query"] = specialty

  list_df.append(page_vacancies_df)

  # суммарное количество страниц
  cnt_pages = query.json()['page']

  # обработка страниц с вакансиями от второй
  # ---------------------------------------
  for specialty in SPECIALTIES:
    for page in range(1, cnt_pages):
      query = requests.get(
                    URL,
                    {"text": specialty,
                     "area": 1,
                     "page": page,
                     "per_page": 100},
                )


  if query.status_code != 200:
    break

  page_df = pd.json_normalize(query.json()["items"])

  page_df["query"] = specialty
  list_df.append(page_df)

vacancies_data = pd.concat(list_df, ignore_index=True)
vacancies_data
````

````python
# удаляю дубли вакансий (поиск дублей по столбцу "id")
vacancies_data = vacancies_data.drop_duplicates(subset='id').reset_index(drop=True)
````

Постройте график `цвет бриллианта - стоимость`:

````python
# собираю сводную таблицу
color_price = diamonds.pivot_table(index='color',
                                       values='total_sales_price',
                                       aggfunc=['min','mean', 'median','max']).rename(columns={'total_sales_price':'tsp'})

color_price.columns = ['_'.join(col).strip() for col in color_price]

color_price.sort_index(inplace=True)
color_price.reset_index(inplace=True)
color_price
````

Отфильтровываю из датафрейма записи с "аномальными"(*) значениями mean_tsp согласно логике:
```
если средняя цена по столбцу mean_tsp меньше:
  mean_tsp < mean(mean_tsp) - 1.5 * std(mean_tsp)
  mean_tsp > mean(mean_tsp) + 1.5 * std(mean_tsp)
то, удалить mean_tsp
```

````python
mean_mean_price, std_mean_price = color_price['mean_tsp'].mean(), color_price['mean_tsp'].std()
under_price, upper_price = mean_mean_price - 1.5 * std_mean_price, mean_mean_price + 1.5 * std_mean_price

filtered_color_price = color_price[(color_price['mean_tsp'] < upper_price) & (color_price['mean_tsp'] > under_price)]
filtered_color_price
````

````python
import plotly.express as px

scatter_fig = px.scatter(
    data_frame=filtered_color_price,
    x='color',
    y='mean_tsp',
    color='color',
    hover_data=['color','min_tsp','mean_tsp', 'median_tsp', 'max_tsp']

)
scatter_fig.update_layout(
    xaxis_title='color name',
    yaxis_title = 'mean price, $',
    title='Price diamonds by colors'
)

scatter_fig.update_traces(
    marker=dict(size=9)
)

scatter_fig.show()
````

````python
carat_price_df = diamonds[['carat_weight', 'total_sales_price']]
carat_price_df.rename(columns={'total_sales_price':'tsp'}, inplace=True)

carat_price_df.loc[:, 'weight_range'] = np.where(
    carat_price_df['carat_weight'] <= 5, 'w1',
    np.where((carat_price_df['carat_weight'] > 5) & (carat_price_df['carat_weight'] <= 10), 'w2',
             np.where((carat_price_df['carat_weight'] > 10) & (carat_price_df['carat_weight'] <= 15), 'w3','w4')
             )
    )

pivot_carat_price_df = carat_price_df.pivot_table(index=['weight_range', 'carat_weight'], values='tsp', aggfunc=['min', 'mean', 'median','max'])
pivot_carat_price_df.columns = ['_'.join(col).strip() for col in pivot_carat_price_df.columns]
pivot_carat_price_df.reset_index(inplace=True)
````

````python
scatter_fig = px.scatter(
    data_frame=pivot_carat_price_df,
    x='weight_range',
    y='median_tsp',
    color='weight_range',
    hover_data=['median_tsp'],

    )

scatter_fig.update_layout(
    xaxis_title='weight range',
    yaxis_title='median sales price, $',
    title={
        'text':'Median salling price ($) by weight range',
        'x':.5,
        'y':.95,
        'xanchor':'center',
        'yanchor':'top',
        'font':{
            'family':'Monospace',
            'size':18,
            'color':'black'
        },
    }
)

scatter_fig.show()
````
